```python
import matplotlib.pyplot as plt
from morai_map_toolkit import MoraiMap
from shapely.geometry import MultiPolygon
```

## Initialize MoraiMap database

```python
morai_map = MoraiMap('./data', 'KATRI')
```

## Map layers
Map은 `drivable_area` 1 가지 레이어로 구성되어 있습니다.

각각의 레이어는 MoraiMap instance의 attribute로 확인이 가능하며 `Dictionary` 구조로 되어 있어 컴포넌트의 토큰으로 접근이 가능합니다.

### Drivable area
Drivable area는 아스팔트로 포장된 영역을 나타내는 polygon layer입니다.

`points`는 exterior polygon을 구성하는 점들의 위치를 나타내며 `interiors`는 외부 polygon 내부에 존재하는 polygon 들의 배열을 나타냅니다.

```python
drivable_area = morai_map.drivable_area
drivable_area_tokens = list(drivable_area.keys())
```

```python
drivable_area[drivable_area_tokens[0]]
```

```python
drivable_area[drivable_area_tokens[0]]['interiors']
```

`MoraiMap` class의 `extract_polygon` 함수를 이용하여 shapely 라이브러리의 Polygon 형태로 추출이 가능합니다.

```python
def draw_drivable_area(polygon):
    plt.plot(*polygon.exterior.xy, color='black')
    for interior in polygon.interiors:
        plt.plot(*interior.xy, color='black')

drivable_area_polygon = morai_map.extract_polygon('drivable_area', drivable_area_tokens[0])

plt.figure(figsize=(12, 24))
plt.axes().set_aspect('equal')

draw_drivable_area(drivable_area_polygon)

plt.show()
plt.close()
```
